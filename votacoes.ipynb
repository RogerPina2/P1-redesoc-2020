## Limpando dataframe com votos individuais

Visualização:

```python
votacoesVotos.head()
```

```python
to_drop = ['uriVotacao', 
           'dataHoraVoto', 
           'deputado_uri', 
           'deputado_nome',
           'deputado_idLegislatura',
           'deputado_urlFoto'
          ]

votacoesVotos.drop(to_drop, inplace=True, axis=1)
```

```python
def limpa_uri(item):
    if isinstance(item, str):
        if 'https' in item:
            return item.split('/')[-1]
        else:
            return item
    else:
        return item
```

```python
votacoesVotos = votacoesVotos.applymap(limpa_uri)
votacoesVotos.rename(columns={ 'deputado_uriPartido' : 'pardido_id' }, inplace=True)
```
