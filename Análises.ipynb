```python
import pandas as pd
```

```python
deputados = pd.read_csv('ArquivosLimpos/deputados.csv')
frentes = pd.read_csv('ArquivosLimpos/frentes.csv')
```

```python
deputados
```

```python
frentes
```

```python
participacao = []
for ids in deputados.deputado_id:
    part = frentes[frentes['deputado_.id'] == ids]
    participacao.append(part.shape[0])
```

```python
deputados['Participação'] = participacao
```
